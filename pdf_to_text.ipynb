```python
from wand.image import Image
import pytesseract
from PIL import Image as PILImage

# Configure the path to the Tesseract executable
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files (x86)\Tesseract-OCR\tesseract.exe"  # Adjust path if needed

# Path to the PDF file
pdf_path = "C:\\Users\\murugan\\Downloads\\sample_input.pdf"

# Convert PDF to images and extract text
all_text = ""
with Image(filename=pdf_path, resolution=300) as pdf:
    for i, page in enumerate(pdf.sequence):
        with Image(page) as img:
            img.format = 'jpeg'  # You can change the format if needed
            image_path = f'page_{i + 1}.jpg'
            img.save(filename=image_path)
            print(f'Page {i + 1} saved as {image_path}')
            
            # Open the saved image and extract text
            with PILImage.open(image_path) as pil_img:
                text = pytesseract.image_to_string(pil_img)
                all_text += f"\n--- Page {i + 1} ---\n{text}\n"

# Print or save the extracted text
print(all_text)

# Optionally, save the output to a text file
with open('extracted_text.txt', 'w', encoding='utf-8') as file:
    file.write(all_text)
```
